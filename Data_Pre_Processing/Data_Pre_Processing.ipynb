```python
import pandas as pd
```

```python
Health = pd.read_csv("health_lka.csv")
```

```python
health = Health[['Year','Indicator Name', 'Value']]
```

```python
health = health.drop(0)
```

```python
health.head()
```

```python
health = health.dropna(subset=['Year', 'Value'])
```

```python
health = health[health['Year'].apply(lambda x: str(x).isdigit())]
health['Year'] = health['Year'].astype(int)
```

```python
health['Indicator Name'] = health['Indicator Name'].astype(str).str.strip()
```

```python
health
```

```python
health['Value'] = pd.to_numeric(health['Value'], errors='coerce')
```

```python
unique_indicators = health['Indicator Name'].unique()
```

```python
indicator_map = {name: f"IND_{str(i+1).zfill(3)}" for i, name in enumerate(unique_indicators)}
```

```python
health['Indicator_Code'] = health['Indicator Name'].map(indicator_map)
```

```python
health
```

```python
health.to_csv("Sri Lanka Health Statistics.csv", index=False)
```
